```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\\Harpreet\\Desktop\\Jupyter_Files\\Heart_Disease_Prediction\\heart.csv")
df.head()
```

```python
df.drop(columns=['Sex'], inplace=True)

# Convert categorical columns to numeric using one-hot encoding
df = pd.get_dummies(df, columns=['ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], drop_first=True)

# Check for and handle any missing data (though in this example there are none)
# You can fill missing values or drop them, depending on your use case


```

```python
correlation_matrix = df.corr()

# Print the correlation matrix
print(correlation_matrix)
```

```python
plt.figure(figsize=(9, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
```

```python
sns.jointplot(x='Age',y='MaxHR',data=df,kind='hex')
```
