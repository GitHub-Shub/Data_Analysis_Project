```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame and it has columns 'Age' and 'Purchase'
# Sample data
df= pd.read_csv("C:\\Users\\Harpreet\Desktop\\Jupyter_Files\\Black_Friday_Analysis\\BlackFriday.csv")

del df['Product_Category_2']
del df['Product_Category_3']



# Group by 'Age' and sum the 'Purchase' column
age_purchase = df.groupby('Age').sum()['Purchase']

# Plot the bar chart
age_purchase.plot(kind='bar', figsize=(12, 6), title='Amount Spend by Age')

# Show the plot
plt.show()
```
