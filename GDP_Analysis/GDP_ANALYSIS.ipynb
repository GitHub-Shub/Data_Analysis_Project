```python

data = df_aw.values
gdp_change = [0]  

for i in range(1, len(data)):
    prev = data[i-1][3]
    curr = data[i][3]
    gdp_change.append(round(((curr - prev) / prev) * 100, 2))

df_aw['GDP'] = gdp_change

df_aw
```

```python
final_data = []

for country_name in df['Country Name'].unique():
    df_aw = df[df['Country Name'] == country_name].copy()  # Copy to avoid SettingWithCopyWarning
    data = df_aw.values
    gdp_change = [0]  

    for i in range(1, len(data)):
        prev = data[i-1][3]
        curr = data[i][3]
        gdp_change.append(round(((curr - prev) / prev) * 100, 2))

    
    if len(gdp_change) == len(df_aw):
        df_aw.loc[:, 'GDP'] = gdp_change
    else:
        raise ValueError("Length of gdp_change does not match the number of rows in df_aw")

    final_data.append(df_aw)


df_aw = pd.concat(final_data)

df_aw
```

```python
import pandas as pd
import plotly.express as px
import plotly.offline as plo

# Filter data for India and China
c1 = df[df['Country Name'] == 'India']
c2 = df[df['Country Name'] == 'China']

# Concatenate the dataframes
df_pr = pd.concat([c1, c2], axis=0)

# Create the plot
df_pr_fig = px.line(df_pr, x='Year', y='Value', title='GDP Comparison', color='Country Name')

# Save the plot to an HTML file
plo.plot(df_pr_fig, filename='Comparisons.html')
```

```python
lst=['IND','ITA','CHN','USA']
dfs=[]
for i in lst:
    dfs.append(df[df['Country Code']== i])
    df_pr=pd.concat(dfs,axis=0)

```

```python
fig=px.line(df_pr,x='Year',y='Value',title='GDP Comparison =' + ''.join(lst),color='Country Name')
plo.plot(fig, filename=''.join(lst)+'.html')
```

```python
def compare_gdp(lst,isOpen):
    dfs=[]
    for i in lst:
          dfs.append(df[df['Country Code']== i])
          df_pr=pd.concat(dfs,axis=0)
    fig=px.line(df_pr,x='Year',y='Value',title='GDP Comparison =' + '  '.join(lst),color='Country Name')
    plo.plot(fig, filename=''.join(lst)+'.html',auto_open=isOpen)
```

```python
compare_gdp(['IND','USA'],False)
```
